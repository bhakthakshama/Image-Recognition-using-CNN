```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

### IMAGE RECOGNITION USING CNN AND TENSORFLOW
---
---

```
# import tensorflow and other libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# useful to visualize plots
```

```
# network parameters
NUM_CLASSES = 10
# reshape 28 X 28 pixels images into a "flatten" array of 784 pixels
RESHAPED = 784
# number of neurons in a hidden layer
HIDDEN_NEURONS = 256
# input image dimensions
IMG_ROWS, IMG_COLS = 28, 28
```

```
# load the fashion mnist dataset
fashion_data = tf.keras.datasets.fashion_mnist
```

```
# split into 60k training and 10k testing samples
(train_images, train_labels), (test_images, test_labels) = fashion_data.load_data()
```

```
train_images = train_images.reshape(60000, RESHAPED)
test_images = test_images.reshape(10000, RESHAPED)
```

```
# to make all pixels value between 0 to 1
train_images, test_images = train_images/255.0, test_images/255.0
```

```
# visualization
# create a dictionary for each type of label
labels = {
    0 : "T-shirt / Top",
    1 : "Trouser",
    2 : "Pullover",
    3 : "Dress",
    4 : "Coat",
    5 : "Sandal",
    6 : "Shirt",
    7 : "Sneaker",
    8 : "Bag",
    9 : "Ankle Boot"
}

plt.figure(figsize = (10, 10))
for i in range(36) : # plotting 36 training images
  plt.subplot(6, 6, i + 1) # arranging in 6 X 6 grid
  plt.xticks([]) # remove xticks
  plt.yticks([]) # remove yticks
  plt.grid(False) # remove grid of the plot
  # below 3 lines mean :
  # take first 36 images and the corresponding labels using the labels dictionary to put the corx rect labels as strings
  plt.imshow(train_images[i].reshape((IMG_ROWS, IMG_COLS)))
  label_index = train_labels[i]
  plt.title(labels[label_index])
plt.show()
```

```
# one hot encoded representation of labels
train_labels = tf.keras.utils.to_categorical(train_labels, NUM_CLASSES)
test_labels = tf.keras.utils.to_categorical(test_labels, NUM_CLASSES)
```

```
# define the architecture of convolutional neural network
# HIDDEN_NEURONS1 = 512
# HIDDEN_NEURONS2 = 256
# HIDDEN_NEURONS3 = 128

class cnn_fmnist :
  @staticmethod
  def build(input_shape, classes) :
    model = Sequential(
        [
            # Dense(HIDDEN_NEURONS1, input_shape = input_shape, name = 'dense_layer_1', activation = 'relu'),
            # Dense(HIDDEN_NEURONS2, input_shape = input_shape, name = 'dense_layer_2', activation = 'relu'),
            # Dense(HIDDEN_NEURONS3, input_shape = input_shape, name = 'dense_layer_3', activation = 'relu'),
            Dense(HIDDEN_NEURONS, input_shape = input_shape, name = 'dense_layer_1', activation = 'relu'),
            Dense(HIDDEN_NEURONS, input_shape = input_shape, name = 'dense_layer_2', activation = 'relu'),
            Dense(classes, name = 'output_layer', activation = 'softmax') # 10 types of clothes
        ] # first layer
    )
    return model
# activation fucntions : softmax for image classfification and sigmoid for email spam detection
```

```
# initialize the optimizer and model
model = cnn_fmnist.build(input_shape = (RESHAPED, ), classes = NUM_CLASSES)

# compile the model
model.compile(optimizer = 'Adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
# SGD => stochastic gradient descent, categorical crossentropy is used whenever softmax is the activation function
# RMSProp is an optimization algorithm that can give faster convergence than SGD because it adapts the step size to different parameters
# and it helps avoid oscillations and slow convergence
# Adam is an extension of SGD and has better adaptation for applications in computer vision and natural language processing due to its
# ability to tackle heavy tail noise
# nadam shows a slight improvement in accuracy over adam

model.summary()
```

```
# training the model
BATCH_SIZE = 128
EPOCHS = 200
VALIDATION_SPLIT = 0.2 # how much of the training set is reserved for validation
```

```
history = model.fit(train_images, train_labels, batch_size = BATCH_SIZE, epochs = EPOCHS, validation_split = VALIDATION_SPLIT)
```

```
# evaluate the model
test_loss, test_acc = model.evaluate(test_images, test_labels)
print("Test accuracy : ", test_acc)
# even after changing the optimizers, we can see that the accuracy is not hitting 90% + border which is because the images are represented
# or considered as flat long vectors so lets try nadam optimizer
```

```
# plot the loss and accuracy functions for the training and validation sets

plt.figure(figsize = (10, 10))
plt.subplot(2, 2, 1)
plt.plot(history.history['loss'], label = 'Loss')
plt.plot(history.history['val_loss'], label = 'Validation Loss')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(history.history['accuracy'], label = 'Accuracy')
plt.plot(history.history['val_accuracy'], label = 'Validation Accuracy')
plt.legend()
plt.title('Train - Accuracy')
```

```
# get the predictions for the test data, convert to an integer between 0 and 9
predicted_classes = np.around(model.predict(test_images))
predicted_classes = np.argmax(predicted_classes, axis = 1)

# get the indices to be plotted
test_labels = np.argmax(test_labels, axis = 1)

correct = np.nonzero(predicted_classes == test_labels)[0]
incorrect = np.nonzero(predicted_classes != test_labels)[0]
```

```
# plot 16 correctly predicted images and their labels

plt.figure(figsize = (15, 15))
for i, idx in enumerate(correct[:16]) :
  plt.subplot(4, 4, i + 1)
  plt.xticks([])
  plt.yticks([])
  plt.grid(False)
  plt.imshow(test_images[idx].reshape((IMG_ROWS, IMG_COLS)), cmap = 'Greens')
  label_index = train_labels[idx]
  plt.title("True : {} Pred : {}".format(labels[test_labels[idx]], labels[predicted_classes[idx]]))
plt.show()
```

```
# plot 16 incorrectly predicted images and their labels

plt.figure(figsize = (15, 15))
for i, idx in enumerate(incorrect[:16]) :
  plt.subplot(4, 4, i + 1)
  plt.xticks([])
  plt.yticks([])
  plt.grid(False)
  plt.imshow(test_images[idx].reshape((IMG_ROWS, IMG_COLS)), cmap = 'Reds')
  label_index = train_labels[idx]
  plt.title("True : {} Pred : {}".format(labels[test_labels[idx]], labels[predicted_classes[idx]]))
plt.show()
```
